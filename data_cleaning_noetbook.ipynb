```python
import pandas as pd
import numpy as np
import os
```

```python
file="./data/census.csv"
```

Read data into pandas dataframe

```python
df = pd.read_csv(file)
df
```

Dataset Overview

```python
df.info()
```

Checking number of na values

```python
df.isna().sum(axis=0)
```

```python
df.describe()
```

```python
categorical_feature_indices = np.where(df.dtypes == object)[0]
categorical_features = df.columns[categorical_feature_indices].to_list()
categorical_features
```

Getting count for each categorical feature value

```python
for feature in categorical_features:
    print("#######################################")
    print("Feature", feature)
    print("#######################################")
    print(df[feature].value_counts())
```

Stripping Extra spaces

```python
df.columns = [c.strip() for c in df.columns]
```

```python
df.info()
```

```python
for feature in categorical_features:
    df[feature.strip()] = df[feature.strip()].str.strip()
```

```python
df
```

Save to csv

```python
df.to_csv("./data/cleaned_data.csv", index=False)
```
